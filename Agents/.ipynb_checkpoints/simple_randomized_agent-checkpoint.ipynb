## PEAS description of the cleaning phase

__Performance Measure:__ Each action costs 1 energy unit. The performance is measured as the sum of the energy units used to clean the whole room.

__Environment:__ A room with $n \times n$ squares where $n = 5$. Dirt is randomly placed on each square with probability $p = 0.2$. For simplicity, you can assume that the agent knows the size and the layout of the room (i.e., it knows $n$). To start, the agent is placed on a random square.

__Actuators:__ The agent can `clean` the current square or move to an adjacent square by going `north`, `east`, `south`, or `west`.

__Sensors:__ Four bumper sensors, one for north, east, south, and west; a dirt sensor reporting dirt in the current square.  


## The agent program for a simple randomized agent

The agent program is a function that gets sensor information (the current percepts) as the arguments. The arguments are:

* A dictionary with boolean entries for the for bumper sensors `north`, `east`, `west`, `south`. E.g., if the agent is on the north-west corner, `bumpers` will be `{"north" : True, "east" : False, "south" : False, "west" : True}`.
* The dirt sensor produces a boolean.

The agent returns the chosen action as a string.

Here is an example implementation for the agent program of a simple randomized agent:  

```python
from numpy import random

actions = ["north", "east", "west", "south", "suck"]

def simple_randomized_agent(bumpers, dirty):
    return random.choice(actions)
```

```python
# define percepts (current location is NW corner and it is dirty)
bumpers = {"north" : True, "east" : False, "south" : False, "west" : True}
dirty = True

# call agent program function with percepts and it returns an action
simple_randomized_agent(bumpers, dirty)
```

__Note:__ This is not a rational intelligent agent. It ignores its sensors and may bump into a wall or not clean a dirty square. You will be asked to implement rational agents below.

## Simple environment example

This simple environment is infinite in size (bumpers are always `False`) and every square is always dirty, even if the agent cleans it. The environment function returns the performance measure which is here the number of cleaned squares (since all squares are constantly dirty, it is the number of `suck` actions by the agent). 

```python
def simple_environment(agent, max_steps, verbose = True):
    num_cleaned = 0
    
    for i in range(max_steps):
        
        dirty = True
        
        bumpers = {"north" : False, "south" : False, "west" : False, "east" : False}

        #asking agent for action
        action = agent(bumpers, dirty)
        if (verbose): print("step", i , "- action:", action) 
        
        if (action == "suck"): 
            num_cleaned = num_cleaned + 1
        
    return num_cleaned
        
```

Do one simulation run with 20 steps.

```python
simple_environment(simple_randomized_agent, max_steps = 20)
```

## Task 1: Implement a simulation environment [2 Points]

The simple environment above is not very realistic. Your environment simulator needs to follow the PEAS description from above. It needs to:

* Initialize the environment by storing the state of each square (clean/dirty) and making some dirty.
* Keep track of the agent's position.
* Call the agent function repeatedly and provide the agent function with the sensor inputs.  
* React to the agent's actions. E.g, by removing dirt from a square or moving the agent around unless there is a wall in the way.
* Keep track of the performance measure. That is, track the agent's actions until all dirty squares are clean and count the number of actions it takes the agent to complete the task.

The easiest implementation for the environment is to hold an 2-dimensional array to represent if squares are clean or dirty and to call the agent function in a loop until all squares are clean or a predefined number of steps have been reached (i.e., the robot runs out of energy).

The simulation environment needs to work with the simple randomized agent program from above and then it can be used for your agent implementation in the tasks below.

```python
#setting up simulation enviroment

import numpy as np
import random

position = [0,0]
def simulation_enviroment(agent, max_steps, n = 5, p = 0.2, verbose = True):
    
    #initializing bumpers
    bumpers = {"north": False, "south": False, "east": False, "west": False}
    
    #setting up 2d array filled w 20% prob of a dirty or cleaned square
    floor = np.random.choice(a=[1, 0], size=(n,n), p =[p, 1-p])
    
    #counting number of dirty squares on the floor
    dirtysquares = np.sum(floor)
    
    #print original floor
    if(verbose):
        print("dirty floor:")
        print(floor)
        
    #if there are dirty squares, how many are there
    if dirtysquares >= 1:
        if(verbose):
            print("dirty squares:", dirtysquares)
        
    #keeping track of agent position & randomizing where the agent starts
    position = [random.randint(0, n-1), random.randint(0, n-1)] #n-1 to set up bounds correctly
    if(verbose):
        print("agent's random starting position", position, "\n")
   
    #performance measure variables to keep track of
    squarescleaned = 0 #initializing to 0 bc no squares have been cleaned yet
    step = 0 #initializing number of steps & starting at step 0
    energy = 0 #initalizing energy consumed
    
    #call the agent function in loop until all squares are clean or max number of steps have been reached
    if (verbose):
        print("agent in action")
    while max_steps <=1000 or dirtysquares > 0:
        
        #setting bumper boundaries aka setting the N,S,W,E walls
        bumpers = {
            "north": position[0] == 0,   #all top 
            "east": position[1] == n-1,  #all right
            "south": position[0] == n-1, #all bottom
            "west": position[1] == 0     #all left
        }
        
        #placing agent in the floor w randomized position
        dirty = floor[position[0],position[1]]
        
        #asking agent for action
        action = agent(bumpers, dirty)
        
        #print current step, action & position
        if (verbose):
            print("step ", step+1 , ": current position", position, "- action:", action)
            
        #react to the agent's actions: if on action, then agent moves position
        if (action == "suck"): #dirty square (1) -> agent cleans up -> now clean square (0)
            floor[position[0],position[1]] = 0  #clean square
            squarescleaned = squarescleaned + 1 #updating the total number of squares agent has cleaned
            
        elif (action == "north"): #moves up by 1 square
            if(position[0] != 0): #checks to make sure that x-pos is not 0 (can't move up if 0,0)
                position[0] = position[0] - 1
            
        elif (action == "east"): #moves right by 1 square
            if(position[1] != n-1): #checks to make sure that y-pos is not 0 (can't move right if 0,4)
                position[1] = position[1] + 1
        
        elif (action == "west"): #moves left by 1 square
            if(position[1] != 0): #checks to make sure that x-pos is not 0 (can't move left if 0,0)
                position[1] = position[1] - 1
            
        elif (action == "south"): #moves down by 1 square
            if(position[0] != n-1): #checks to make sure that y-pos is not 0 (can't move down if 4,0)
                position[0] = position[0] + 1
                
        #updating number of iteration/step by 1
        step = step + 1
        
        #updating the performance measure
        energy = energy + 1
                
        #updating the number of dirty squares
        dirtysquares = np.sum(floor)
        
        #updating the number of max steps & decreasing it by 1 (aka 1 iteration has occured)
        max_steps = max_steps - 1
        
        #break the loop when no more dirty squares or max num of steps is reached
        if dirtysquares == 0 or max_steps == 0:
            break
            
    #print new floor & stats
    if(verbose):
        print("\nnew floor:")
        print(floor)
        print("dirty squares:", dirtysquares, "\ntotal steps:", step)
        print("performance measure:", energy, "energy units")
    
    #returns the number of total steps
    return step
```

```python
simulation_enviroment(simple_randomized_agent, max_steps = 1000)
```

## Task 2:  Implement a simple reflex agent [1 Point] 

The simple reflex agent randomly walks around but reacts to the bumper sensor by not bumping into the wall and to dirt with sucking. Implement the agent program as a function.

_Note:_ The agent cannot directly use variable in the environment. It only gets the percepts as the arguments to the agent program function.

```python
# look at percepts & then have a set of rules to decide what to do 
# doesn't have memory, so can't remember past actions -> random action
# simple reflex agent- if dirty, it sucks or travels randomly

import numpy as np
import random

def simple_reflex(bumpers, dirty):
    
    #possible directions
    direction = ["north", "east", "west", "south"]
    
    #if dirty -> clean
    if(dirty):
        return "suck"
    
    #if agent hits north wall -> random action
    if (bumpers["north"] == True):
        direction.remove("north") #removes north from list of possible action
        return np.random.choice(direction) #returns random possible actions: "east", "west", "south"
    
    #if agent hits east wall -> random action
    if (bumpers["east"] == True):
        direction.append("north") #adds north back to the list 
        direction.remove("east") #removes east from list of possible action
        return np.random.choice(direction) #returns random possible actions: "west", "south", "north"
    
    #if agent hits west wall -> random action
    if (bumpers["west"] == True):
        direction.append("east") #adds east back to the list 
        direction.remove("west") #removes west from list of possible action
        return np.random.choice(direction) #returns random possible actions: "south", "north", "east"
    
    #if agent hits south wall -> random action
    if (bumpers["south"] == True):
        direction.append("west")
        #adds west back to the list 
        direction.remove("south") #removes south from list of possible action
        return np.random.choice(direction) #returns random possible actions: "north", "east", "west"
    
    #re-setting all possible actions
    direction = ["north", "east", "west", "south"]
    
    #returning random possible actions
    return np.random.choice(direction)
```

```python
simulation_enviroment(simple_reflex, max_steps = 1000)
```

## Task 3: Implement a model-based reflex agent [3 Point]

This agent keeps track of the location and remembers where it has cleaned. Assume the agent knows how many squares the room has. It can move to a corner to determine its location and then is able to use more advanced navigation.

Describe how you define the __agent state__ and how your agent works before implementing it. _Note on implementing the state in Python:_ [Examples](https://github.com/mhahsler/CS7320-AI/blob/master/Python_Code_Examples/store_agent_state_information.ipynb)

```python
#TO DO
#Your short description of the state and your implementation goes here


# take our simple reflex agent and give it some memory(aka state information: remember what it did before) 
# to do better things such as navigation but using only sensors we have (bumper and dirty)
# describe how to navigate a room 
```

```python
# model based reflex agent

import numpy as np
import random

#global state variables 

#keeping track of square agent has visited
#empty bc agent has not visited any squares yet 
visitedsquare = []

#keeping track if agent is heading east/right
#prefixed state of agent heading east
#if agent needs to head east -> true
#if agent needs to head west -> false
right = True

#function
def model_based_reflex(bumpers, dirty):
        
    #global functions
    global visitedsquare
    global right
    
    #keep track of agent's current position
    #0,0 bc agent will always start at NW corner
    currpos = [0,0]
    
    #agent has not visited any squares
    if(len(visitedsquare) == 0):
        
        #if agent is at 0,0 -> north and west bumpers: activated
        if(bumpers["north"] == True and bumpers["west"] == True):
            
            #adding current agent's current position to visitedsquare list aka adding 0,0 pos to list
            #self-note: different position variable than the one from enviroment, 
            #since agent & enviroment cannot share any variables  
            visitedsquare.append(currpos)
            
            #if square dirty -> clean
            if(dirty):
                return "suck"
            
            #agent at north wall -> move 1 unit to the right
            if(bumpers["north"] == True):
                
                #aka adding 0,1 currpos to the list
                visitedsquare.append( [currpos[0],currpos[1]+1] ) #adding new agent position to visitedsquares
                return "east"
            
            #agent at north & east wall -> so move down by 1 unit
            visitedsquare.append( [currpos[0]+1,currpos[1]] ) #adding new agent position to visitedsquares
            return "south"
        
        #north=false and west=false meaning, agent is not at 0,0
        elif(bumpers["north"] != True and bumpers["west"] != True): 
            
            #move agent to 0,0 by either going north or west
            directions = ["north", "west"]
            return np.random.choice(directions)
       
        #if agent hits north wall -> agent travels to west wall
        elif(bumpers["north"] == True): 
            return "west"
        
        #if agent hits west wall -> agent travels to north wall
        return "north"
        
        
    #agent has visited 1 square already
    else:
        
        #if dirty -> suck 
        #self-note: current pos not added bc agent will not move after "suck"
        if(dirty):
            return "suck"
        
        #agent is heading east
        if(right == True):
            
            #agent hit east wall
            if(bumpers["east"] == True):
                
                #can no longer move right, agent needs to move left, 
                right = False
                
                #want agent to move down by 1 square
                visitedsquare.append( [currpos[0]+1,currpos[1]] ) #adding new agent position to visitedsquares
                return "south"
            
            #move right by 1 square
            visitedsquare.append( [currpos[0],currpos[1]+1] ) #adding new agent position to visitedsquares
            return "east"
        
        #agent can no longer go east bc it hit east wall -> agent needs to move  west
        if(right == False):
            
            #agent hits east wall
            if(bumpers["east"] == True):
                
                #agent moves west by 1
                visitedsquare.append( [currpos[0],currpos[1]-1] ) #adding new agent position to visitedsquares
                return "west"
            
            #agent hits west wall
            if(bumpers["west"] == True):
                
                #can no longer move left -> agent needs to move right
                right = True
                
                #go down by 1
                visitedsquare.append( [currpos[0]+1,currpos[1]] ) #adding new agent position to visitedsquares
                return "south"
            
            #agent moves west by 1 square
            visitedsquare.append( [currpos[0],currpos[1]-1] ) #adding new agent position to visitedsquares
            return "west"
```

```python
simulation_enviroment(model_based_reflex, max_steps = 1000)
```
